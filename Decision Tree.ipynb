## 2. 클래스 분리

```python
X = data.drop(['target'], axis=1)
y = data['target']
```

## 3. 데이터학습

```python
dt_clf = DecisionTreeClassifier()
```

- 로드된 의사결정 트리 분류 모듈을 변수 dt_clf에 저장한다.

```python
dt_clf = dt_clf.fit(X_train, y_train)
```

- 의사결정 트리 분류 모듈이 저장된 변수 dt_clf 의 함수 fit() 에 변수 X_train, y_train 을 입력해 의사결정 트리 분류 모델을 생성한다.
- 그런 다음 생성한 모델을 다시 변수 dt_clf에 저장한다.

```python
dt_prediction = dt_clf.predict(X_test)
```

- 변수 dt_clf의 함수 predict()에 변수 X_test를 입력한다. 
- 그리고 입력한 X_test에 대한 클래스 예측 값을 변수 dt_prediction에 저장한다.

## 4. 의사결정트리 그래프 표현

```python
os.environ["PATH"] += os.pathsep + 'C:\Users\lucky\anaconda3\Lib\site-packages\pydotplus'
os.environ['PATH'] = os.environ['PATH']+';'+os.environ['CONDA_PREFIX']+r"\Lib\site-packages\pydotplus\graphviz"
```

- ipython 내에서 그래프를 생성할 수 있는 인터페이스 경로를 추가 설정하는 부분

```python
feature_names = data.columns.tolist()
```

- 트리표현 함수에 입력되는 파라미터 중 하나인 feature_names에 값을 입력하기 위해, 변수 tennis_data의 각 컬럼명을 list형태로 변환한 후 변수 feature_names에 저장한다.

```python
feature_names = feature_names[0:365]
```

```python
target_name = np.array(['고', '중', '저'])
```

- 트리표현 함수에 입력되는 파라미터 중 하나인 class_name에 값을 입력하기 위해, Target_class 값인 '고', '중', '저'를 배열형태로 변수 target_name에 저장한다.

```python
dt_dot_data = tree.export_graphviz(dt_clf, out_file = None,
                                  feature_names = feature_names,
                                  class_names = target_name,
                                  filled = True, rounded = True,
                                  special_characters = True)
```

```python
dt_graph = pydotplus.graph_from_dot_data(dt_dot_data)
```
